```python
import random

random.seed()

precioCompra = 800 # Por kilo
precioVenta = 1500 # Por kilo
utilidad = precioVenta - precioCompra
utilidadPerdida = 0
inventario = 0
dias = 365
cantidadOptima = 100
utilidadTotal = 0

for dia in range(dias):
  probabilidadVenta = random.random()
  probabilidadPerdida = random.random()

  perdidaInventario = 0
  donacionDiaria = 0

  print(f'Dia {dia + 1}')
  print(f'Inventario al inicio del dia: {inventario} kilos')

  # Posibilidad de perder inventario
  if (probabilidadPerdida < 0.45):
    perdidaInventario = precioCompra * inventario
    inventario = 0

  print(f'Inventario luego de perdidos: {inventario} kilos')

  if (probabilidadVenta < 0.25):
    cantidadVendida = 5
  elif (probabilidadVenta < 0.40):
    cantidadVendida = 10
  elif (probabilidadVenta < 0.75):
    cantidadVendida = 12
  elif (probabilidadVenta < 0.875):
    cantidadVendida = 15
  else:
    cantidadVendida = 20

  cantidadSinCubrir = cantidadVendida * 10 - (cantidadOptima + inventario)

  utilidadPerdida = cantidadSinCubrir * utilidad

  # Utilidad diaria
  if (utilidadPerdida > 0):
    utilidadDiaria = ((cantidadVendida * 10) * utilidad) - perdidaInventario - utilidadPerdida
  else:
    utilidadDiaria = ((cantidadVendida * 10) * utilidad) - perdidaInventario

  # Donación
  if (cantidadVendida * 10 < cantidadOptima + inventario):
    cantidadDonar = int((cantidadVendida * 10) / 10) * 2
    if (cantidadDonar <= (cantidadOptima + inventario - cantidadVendida * 10)):
      donacionDiaria = cantidadDonar
    else:
      donacionDiaria = (cantidadOptima + inventario - cantidadVendida * 10)
  else:
    donacionDiaria = 0

  # Inventario
  if (cantidadVendida * 10 < cantidadOptima + inventario):
    inventario += cantidadOptima - donacionDiaria - (cantidadVendida * 10)
  else:
    inventario = 0

  utilidadTotal += utilidadDiaria

  print(f'Cantidad optima: {cantidadOptima} kilos')
  print(f'Cantidad vendida: {cantidadVendida * 10} kilos')
  print(f'Cantidad sin cubrir: {cantidadSinCubrir} kilos')
  print(f'Utilidad perdida: {utilidadPerdida} kilos')
  print(f'Donación diaria: {donacionDiaria} kilos')
  print(f'Utilidad diaria: ${utilidadDiaria}')

print(f'Utilidad total: ${utilidadTotal}')



```
