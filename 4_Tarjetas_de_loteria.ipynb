```python
import random

random.seed()

tarjetas = 100000
cantidadGanadores = 0
cantidadPerdedores = 0
precioTarjeta = 2500
gananciaTotal = 0
gananciaSinGasto = 0
gastoTotal = 0
valoresBoleta = (1000, 5000)
filas = 3

def check(valores):
    return all(i == valores[0] for i in valores)

for tarjeta in range(tarjetas):
  valoresSelecionados = []

  for fila in range(filas):
    probabilidadValor = random.random()

    if (probabilidadValor < 0.5):
      valoresSelecionados.append(valoresBoleta[0])
    else:
      valoresSelecionados.append(valoresBoleta[1])

  gano = check(valoresSelecionados)
  suma = 0

  if (gano):
    cantidadGanadores += 1
    suma = sum(valoresSelecionados)
  else:
    cantidadPerdedores += 1

  print(f'Tarjeta {tarjeta + 1} Resultado: {gano}')
  print(f'Valores: {valoresSelecionados}')
  print(f'Valor ganado: {suma}')

  gastoTotal += suma

gananciaSinGasto = tarjetas * precioTarjeta
gananciaTotal = gananciaSinGasto - gastoTotal

print(f'\nGanancia sin gasto: {gananciaSinGasto}')
print(f'Gasto total: {gastoTotal}')
print(f'Ganancia total: {gananciaTotal}')
print(f'Cantidad de ganadores: {cantidadGanadores}')
print(f'Cantidad de perdedores: {cantidadPerdedores}')











```
