```python
import random

random.seed()

dias = 100
semanas = round(dias / 7, 3)
produccionDiaria = 50
costoTorta = 1000
precioTorta = 3000
utilidad = precioTorta - costoTorta
multa = 30000
permiso = 20000
pagoPermisos = round(semanas, 3) * permiso
tortasVendidas = 0
tortasNoVendidas = 0
ganancia = 0
multas = 0

for dia in range(dias):
  probabilidadDemanda = random.random()

  print(f'Dia {dia + 1}')

  if (probabilidadDemanda < 0.1):
    demanda = 10
  elif (probabilidadDemanda < 0.3):
    demanda = 20
  elif (probabilidadDemanda < 0.6):
    demanda = 25
  elif (probabilidadDemanda < 0.8):
    demanda = 30
  elif (probabilidadDemanda < 0.9):
    demanda = 50
  elif (probabilidadDemanda < 0.96):
    demanda = 70
  else:
    demanda = 100

  tortas = 0

  if (demanda > 50):
    tortas = 50
  else:
    tortas = demanda
    tortasNoVendidas += produccionDiaria - demanda

  tortasVendidas += tortas

  print(f'Demanda de tortas: {demanda}')
  print(f'Tortas vendidas: {tortas}')

  probabilidadMulta = random.random()

  if (probabilidadMulta < 0.25):
    multas += 1

ganancia = (tortasVendidas * utilidad)

print(f'\nCantidad media de tortas vendidas por dia: {tortasVendidas / dias}')
print(f'Cantidad media de tortas no vendidas por dia: {tortasNoVendidas / dias}')
print(f'Cantidad media de tortas a botar por dia: {tortasNoVendidas / dias}')
print(f'Utilidad media por dia: ${ganancia / dias}')
print(f'Cantidad de multas: {multas}')
print(f'Valor total a pagar por multas: ${multas * multa}')
print(f'Ganancias totales: ${ganancia}')

print(f'\nValor de permisos por las {semanas} semanas: ${pagoPermisos}')

if (pagoPermisos < multas * multa):
  print('Se recomienda pagar los permisos')
else:
  print('No se recomienda pagar los permisos')

```
