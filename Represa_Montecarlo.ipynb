```python
import random

random.seed()

unidadesAcumuladas = 1
anios = 15
meses = 12
unidadeDesechadas = 0
unidadesBajas = 0

for anio in range(anios):
  print(f'Año: {anio + 1}')

  for mes in range(meses):
    probabilidad = random.random()
    cantidad = 0

    if (mes != 0):
      if (probabilidad < 0.15):
        cantidad = 0
      elif (probabilidad < 0.50):
        cantidad = 1
      elif (probabilidad < 0.80):
        cantidad = 2
      else:
        cantidad = 3

    unidadesAcumuladas += cantidad

    print(f'Mes: {mes + 1} - Unidades sin ajustar:  {unidadesAcumuladas}')

    if (unidadesAcumuladas > 4):
      unidadeDesechadas += unidadesAcumuladas - 4
      unidadesAcumuladas = 4

    print(f'Mes: {mes + 1} - Unidades ajustadas:  {unidadesAcumuladas}')

    if (unidadesAcumuladas == 1):
      unidadesBajas += 1

    if (unidadesAcumuladas > 2):
      unidadesAcumuladas -= 2
    elif (unidadesAcumuladas == 1):
      unidadesAcumuladas -= 1

print(f'Cantidad de unidades desechadas:  {unidadeDesechadas}')
print(f'Cantidad de generación de energia con una unidad:  {unidadesBajas}')
```
